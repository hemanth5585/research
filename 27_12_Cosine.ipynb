```python
#Cosine similarity on rows of a matrix
import scipy.spatial as sp
import pandas as pd
```

```python
df = pd.read_excel("/content/3tuple (1).xlsx")
df
```

```python
n = len(df.index)
mat =[]
for i in range(n):
  print(i)
  a = df.iloc[i].tolist()
  mat.append(a)
print(mat)
```

```python
fin = []
for i in mat:
  x = i
  x.pop(0)
  fin.append(x)
for i in fin:
  print(len(i))
```

```python
cos_mat = []
for i in fin:
  lis = []
  for j in fin:
    x =[]
    y = []
    if i==j:
      val = 0
    else:
      #print(i)
      #print(j) 
      x.append(i)
      y.append(j)
      val = sp.distance.cdist(x,y,'cosine').tolist()
      val = val[0][0]
    lis.append(val)
  cos_mat.append(lis)
print(cos_mat)
```

```python
for i in cos_mat:
  print(i)
```

```python
topic_list = df.iloc[:,0]
topic_list =topic_list.tolist()
print(topic_list) 
df1 = pd.DataFrame(cos_mat)
df1.index = topic_list
df1.head()
```
